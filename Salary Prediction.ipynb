```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
##Dataset Preparation
data = pd.read_csv('Salary_Data.csv')
```

```python
data.head(10)
```

```python
#Extract the Years of Experiece and Salary by using variables
x = data.iloc[:,0:1].values
y = data.iloc[:,1].values
```

```python
#Plotting data of x and y
sns.barplot(x='YearsExperience' , y='Salary' , data = data)
```

```python
#Representing the relation of x and y using heatmap
print(data.corr())
sns.heatmap(data.corr())
```

```python
# Splitting data into train dataSet for our regression model
## We are going to use test size of 1/3
from sklearn.model_selection import train_test_split
x_train, x_test, y_train , y_test = train_test_split(x,y, test_size = 1/3, random_state = 0)
```

### Now we will train and test the model

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
#fitting the data
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
y_pred
```

#Result Visualization


```python
#Model for Training Data
plt.scatter(x_train,y_train,color = 'red')
plt.plot(x_train, lr.predict(x_train), color='blue')
plt.title('Salary vs Years of Experience(training data)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary of an Employee')
```

```python
#Model for Testing Data
plt.scatter(x_test,y_test,color = 'red')
plt.plot(x_train, lr.predict(x_train), color='blue')
plt.title('Salary vs Years of Experience(testing data)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary of an Employee')
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
#We have the above percentage close to 1 so model is perfectly fit
```

```python
diff_LR=y_test-y_pred
```

```python
res_data=pd.concat([pd.Series(y_pred),pd.Series(y_test),pd.Series(diff_LR)],axis=1)
res_data.columns=['Prediction','Original Data','Diff']
```
